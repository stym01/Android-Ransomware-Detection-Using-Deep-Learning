```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("data.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
data.info()
```

```python
features=data.drop("label",axis=1)
```

```python
targets=data["label"]
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(features,targets,test_size=0.2,random_state=42,stratify=targets)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

**Now using Random Forest**

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
accuracy_score(Y_test,rfc.predict(X_test))
```

```python
precision_score(Y_test,rfc.predict(X_test))
```

```python
recall_score(Y_test,rfc.predict(X_test))
```

```python
f1_score(Y_test,rfc.predict(X_test))
```
